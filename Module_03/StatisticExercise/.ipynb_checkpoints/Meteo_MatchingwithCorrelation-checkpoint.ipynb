```python
import numpy as np
import math
import cv2
from google.colab.patches import cv2_imshow
```

```python
def computeMean (X):
    return np.sum(X)/len(X)
X = [2, 0, 2, 2, 7, 4, -2, 5, -1, -1]
print (" Mean : ", computeMean (X))
print(np.mean(X))
```

```python
def computeMedian (X):
    size = len(X)
    X = np.sort(X)
    print (X)
    if ( size % 2 == 0):
        indexcenter = int(size/2)
        return (1/2)*(X[indexcenter]+X[indexcenter-1])
    else:
        indexcenter = int(size/2)
        return X[indexcenter]

X = [1, 5, 4, 4, 9, 13]
print (" Median : ", computeMedian (X))
```

```python
def compute_std (X):
    mean = computeMean (X)
    variance = 0
    for i in range(len(X)):
        variance = variance + (X[i]-mean)**2
    variance = variance/len(X)
    return np.sqrt( variance )
X = [ 171 , 176 , 155 , 167 , 169 , 182]
print(np.round(compute_std(X),2))
```

```python
def computeCorrelationCofficient (X, Y):
    N = len(X)
    numerator = 0
    denominator = 0
    product = X.dot(Y)
    numerator = N*product - np.sum(X)*np.sum(Y)
    denominator = math.sqrt(N*np.sum(X**2) - np.sum(X)**2) * math.sqrt(N*np.sum(Y**2) - np.sum(Y)**2)
    return numerator/denominator

X = np. asarray ([-2, -5, -11, 6, 4, 15, 9])
Y = np. asarray ([4 , 25, 121 , 36, 16, 225 , 81])
print (" Correlation : ", computeCorrelationCofficient (X,Y))
```

```python
def cosine_similarity(x,y):
    numerator = np.dot(x,y)
    denominator = np.linalg.norm(x) * np.linalg.norm(y)
    return numerator/denominator
```

```python
def computeSAD(wp_flattened,wqd_flattened):
    return sum(abs(wp_flattened-wqd_flattened))
```

```python
def window_based_matching_SAD(left_img,right_img,disparity_range,kernel_size= 5):
    left = cv2.imread(left_img,0)
    right= cv2.imread(right_img,0)

    left = left.astype(np.float32)
    right = right.astype(np.float32)

    height, width = left.shape[:2]

    depth = np.zeros((height,width),np.uint8)
    kernel_half = int((kernel_size-1)/2)
    scale = 3

    for y in range(kernel_half,height-kernel_half):
        print("\rProcessing ... %d%% complete"%(y/(height-kernel_half)*100),end="",flush=True)
        for x in range(kernel_half,width-kernel_half):
            disparity = 0
            cost_optimal = 999999

            for j in range(disparity_range):
                d= x-j
                cost = -1
                if (d-kernel_half) >0:
                    wp = left[(y-kernel_half):(y+kernel_half)+1, (x-kernel_half):(x+kernel_half)+1]
                    wqd = right[(y-kernel_half):(y+kernel_half)+1, (d-kernel_half):(d+kernel_half)+1]

                    wp_flattened = wp.flatten()
                    wqd_flattened = wqd.flatten()
                    cost = computeSAD(wp_flattened,wqd_flattened)

                if cost < cost_optimal :
                    cost_optimal = cost
                    disparity = j

            depth[y,x] = disparity* scale

    return depth
```

```python
left_img_path = "/content/drive/MyDrive/AI_Course/Module_03/230730 - Statistic Exercise - Dr_Dinh_Vinh/MiddleburyDataset/normal condition/view1.png"
right_img_path = "/content/drive/MyDrive/AI_Course/Module_03/230730 - Statistic Exercise - Dr_Dinh_Vinh/MiddleburyDataset/normal condition/view5.png"
disparity_range = 64
kernel_size = 5
depth = window_based_matching_SAD(
    left_img_path,
    right_img_path,
    disparity_range,
    kernel_size=kernel_size)

cv2_imshow(depth)
cv2_imshow(cv2.applyColorMap(depth,cv2.COLORMAP_JET))
```

```python
def window_based_matching_cosine(left_img,right_img,disparity_range,kernel_size= 5):
    left = cv2.imread(left_img,0)
    right= cv2.imread(right_img,0)

    left = left.astype(np.float32)
    right = right.astype(np.float32)

    height, width = left.shape[:2]

    depth = np.zeros((height,width),np.uint8)
    kernel_half = int((kernel_size-1)/2)
    scale = 3

    for y in range(kernel_half,height-kernel_half):
        print("\rProcessing ... %d%% complete"%(y/(height-kernel_half)*100),end="",flush=True)
        for x in range(kernel_half,width-kernel_half):
            disparity = 0
            cost_optimal = -1

            for j in range(disparity_range):
                d= x-j
                cost = -1
                if (d-kernel_half) >0:
                    wp = left[(y-kernel_half):(y+kernel_half)+1, (x-kernel_half):(x+kernel_half)+1]
                    wqd = right[(y-kernel_half):(y+kernel_half)+1, (d-kernel_half):(d+kernel_half)+1]

                    wp_flattened = wp.flatten()
                    wqd_flattened = wqd.flatten()
                    cost = cosine_similarity(wp_flattened,wqd_flattened)

                if cost > cost_optimal :
                    cost_optimal = cost
                    disparity = j

            depth[y,x] = disparity* scale

    return depth
```

```python
depth_cosine = window_based_matching_cosine(
    left_img_path,
    right_img_path,
    disparity_range,
    kernel_size=kernel_size)

cv2_imshow(depth_cosine)
cv2_imshow(cv2.applyColorMap(depth_cosine,cv2.COLORMAP_JET))
```

```python
def compute_rms_error(matrix1,matrix2):
  return np.sqrt(np.sum(np.square(matrix1-matrix2))/(matrix1.shape[0]*matrix1.shape[1]))

groundtruth_img_path = "/content/drive/MyDrive/AI_Course/Module_03/230730 - Statistic Exercise - Dr_Dinh_Vinh/MiddleburyDataset/disp1.png"
groundtruth_image = cv2.imread(groundtruth_img_path,0)
rmse = compute_rms_error(depth_cosine , groundtruth_image )
print(rmse)
```

```python
def window_based_matching_correlation(left_img,right_img,disparity_range,kernel_size= 5):
    left = cv2.imread(left_img,0)
    right= cv2.imread(right_img,0)

    left = left.astype(np.float32)
    right = right.astype(np.float32)

    height, width = left.shape[:2]

    depth = np.zeros((height,width),np.uint8)
    kernel_half = int((kernel_size-1)/2)
    scale = 3

    for y in range(kernel_half,height-kernel_half):
        print("\rProcessing ... %d%% complete"%(y/(height-kernel_half)*100),end="",flush=True)
        for x in range(kernel_half,width-kernel_half):
            disparity = 0
            cost_optimal = -1

            for j in range(disparity_range):
                d= x-j
                cost = -1
                if (d-kernel_half) >0:
                    wp = left[(y-kernel_half):(y+kernel_half)+1, (x-kernel_half):(x+kernel_half)+1]
                    wqd = right[(y-kernel_half):(y+kernel_half)+1, (d-kernel_half):(d+kernel_half)+1]

                    wp_flattened = wp.flatten()
                    wqd_flattened = wqd.flatten()
                    cost = computeCorrelationCofficient(wp_flattened,wqd_flattened)

                if cost > cost_optimal :
                    cost_optimal = cost
                    disparity = j

            depth[y,x] = disparity* scale

    return depth
```

```python
depth_correlation = window_based_matching_correlation(
    left_img_path,
    right_img_path,
    disparity_range,
    kernel_size=kernel_size)

cv2_imshow(depth_correlation)
cv2_imshow(cv2.applyColorMap(depth_correlation,cv2.COLORMAP_JET))
rmse_correlation = compute_rms_error(depth_correlation , groundtruth_image )
print(rmse_correlation)
```
