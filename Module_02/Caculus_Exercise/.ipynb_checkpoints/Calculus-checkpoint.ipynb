```python
import math
import matplotlib . pyplot as plt
import numpy as np
import cv2
```

```python
# Exponential f(x) = e^x
def exponential(x):
    return math.e**x
def derivative_exponential(x):
    return exponential(x)

def plot_function_and_derivative (x, f_x , df_x ):
    
    fig , ax = plt.subplots( figsize =(10 , 5))
    ax.spines['left'].set_position('center')
    ax.spines['bottom'].set_position('zero')
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.yaxis.set_ticks_position('left')

    # plot the function
    ax.plot(x, f_x, color ="#d35400", linewidth =3, label ='function')
    ax.plot(x, df_x, color ="#1abd15", linewidth =3, label ='derivative' )
    ax.legend (loc="upper left", frameon = False )
    plt.show()

x_data = list(np.arange(-5,5,0.1))
f_x = [exponential(x) for x in x_data]
df_x=[derivative_exponential(x) for x in x_data]
plot_function_and_derivative(x_data,f_x,df_x)
```

# PReLU function α = 0.25, f(x)=αx if x<=0, , x if x>0

```python
def PReLU(x):
    if x<=0:
        return 0.25*x
    else:
        return x
def derivative_PReLU(x):
    if x<=0:
        return 0.25
    else:
        return 1
x_data = list(np.arange(-50,50,1))
f_x = [PReLU(x) for x in x_data]
df_x=[derivative_PReLU(x) for x in x_data]
plot_function_and_derivative(x_data,f_x,df_x)    
```

# ELU function: α = 0.25, f(x) = α*(e^x − 1),x<=0
#                                                =x                , x>0


```python
def ELU(x):
    if x<=0:
        return 0.25*(math.e**x - 1)
    else :
        return x
def derivative_ELU(x):
    if x<=0:
        return ELU(x)+0.25
    else:
        return 1
x_data = list(np.arange(-5,5,0.001))
f_x = [ELU(x) for x in x_data]
df_x=[derivative_ELU(x) for x in x_data]
plot_function_and_derivative(x_data,f_x,df_x)  
```

# Softplus function: f(x) = log(1 + e^x)

```python
def softplus(x):
    return math.log10(1+math.e**x)
def derivative_softplus(x):
    return 1/(1+math.e**(-x))

x_data = list(np.arange(-5,5,0.001))
f_x = [softplus(x) for x in x_data]
df_x=[derivative_softplus(x) for x in x_data]
plot_function_and_derivative(x_data,f_x,df_x)  
```

# Softsign function f(x) = x /(|x| + 1)

```python
def softsign(x):
    return x/(abs(x)+1)
def derivative_softsign(x):
    return 1/(1+abs(x))*2

x_data = list(np.arange(-100,100,1))
f_x = [softsign(x) for x in x_data]
df_x=[derivative_softsign(x) for x in x_data]
plot_function_and_derivative(x_data,f_x,df_x)  
```

# def sigmoid (x) f(x) = 1/1 + e**(-x)

```python
def sigmoid (x):
    return 1 / (1 + np.exp(-x))
def sigmoid_list ( x_data ):
    fx = [ sigmoid (x) for x in x_data ]
    return fx
def sigmoid_derivative (x):
    y = sigmoid (x)
    return y*(1 -y)
def d_sigmoid_list ( x_data ):
    dfx = [ sigmoid_derivative (x) for x in x_data ]
    return dfx
x_list = list ([ -1, 0, 1])
x_sigmoid_list = sigmoid_list ( x_list )
print ([ round (num , 2) for num in x_sigmoid_list ])
x_list = list ([ -1, 0, 1])
x_gradient_list = d_sigmoid_list ( sigmoid_list ( x_list ))
print ([ round (num , 2) for num in x_gradient_list ])
```

# tìm xấp xỉ x mà f(x) (ví dụ f(x) = 3x4 − 4x2 − 6x − 3) là minimum.

```python
def computeXDerivative ( image ):
    w = len (image[0])
    h = len (image)
    x_derivative = [[0]* w for _ in range (h)]
#     x_derivative = np.zeros((h,w))
    for i in range(0,h):
        for j in range(1,w-1):
#             print(f'{i} {j}')
            value = image[i][j+1]-image[i][j-1]
            value = value+127.5
            value = min(value,255)
            value = max(value,0)
            
            x_derivative[i][j] = value
            if(value>255 or value<0):
                print(f'{i} {j} {value}')
    return x_derivative
def computeYDerivative ( image ):
    w = len( image[0])
    h = len( image)
    y_derivative = [[0]* w for _ in range (h)]
#     y_derivative = np.zeros((h,w))
    for i in range(0,w):
        for j in range(1,h-1):
            value = image[j+1][i]-image[j-1][i]
            value = value+127.5
            value = min(value,255)
            value = max(value,0)
            y_derivative[j][i] = value  
            if(value>255 or value<0):
                print(f'{i} {j} {value}')

    return y_derivative
def computeXYDerivative ( image ):
    w = len( image[0])
    h = len( image)
    xy_derivative = [[0]* w for _ in range (h)]
#     xy_derivative = np.zeros((h,w))
    x_derivative= computeXDerivative(image)
    y_derivative=computeYDerivative(image)
    for i in range(0,w):
        for j in range(0,h):
            value = math.sqrt(x_derivative[i][j]**2 + y_derivative[i][j]**2)
            value = min(value,255)
            value = max(value,0)
            xy_derivative[i][j]= value
            if(value>255 or value<0):
                print(f'{i} {j} {value}')

    return xy_derivative


cat_image = cv2.imread('cat.jpeg',0)
cat_image = cv2.resize(cat_image, (400 ,400), interpolation = cv2.INTER_AREA ).tolist()
print(type(cat_image))
x_derivative =computeXDerivative(cat_image)
# x_derivative = np.array(computeXDerivative(cat_image))

print(type(x_derivative))
y_derivative =computeYDerivative(cat_image)
# y_derivative =  np.array(computeYDerivative(cat_image))
xy_derivative= computeXYDerivative(cat_image)
# xy_derivative = np.array(computeXYDerivative(cat_image))
# create figure
fig = plt.figure( figsize =(10 , 7))

# showing image
fig.add_subplot (2, 2, 1)
plt.imshow( cat_image , cmap ="gray")
plt.axis ('off')
plt.title (" Input image ")

fig.add_subplot (2, 2, 2)
plt.imshow( x_derivative , cmap ="gray")
plt.axis('off')
plt.title(" Gradient in X- direction ")

fig.add_subplot(2, 2, 3)
plt.imshow( y_derivative , cmap ="gray")
plt.axis('off')
plt.title(" Gradient in Y- direction ")

fig.add_subplot(2, 2, 4)
plt.imshow( xy_derivative , cmap ="gray")
plt.axis ('off')
plt.title ("Gradient Magnitude ")
```
