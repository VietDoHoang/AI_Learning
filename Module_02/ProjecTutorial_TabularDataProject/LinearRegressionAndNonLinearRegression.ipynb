```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class LinearRegression:
    def __init__(self, X_data, y_target, learning_rate=0.01, num_epochs=1000):
        self.X_data        = X_data
        self.y_target      = y_target
        self.learning_rate = learning_rate
        self.num_epochs    = num_epochs
        self.num_samples   = self.X_data.shape[0]
        
        self.theta         = np.random.randn(self.X_data.shape[1])
        self.losses        = []
    
    def compute_loss(self, y_pred, y_target):
        loss = (y_pred-y_target)**2
        loss = np.mean(loss)
        return loss
    
    def predict(self, X_data):
        y_pred = X_data.dot(self.theta)
        return y_pred
    
    def fit(self):
        for epoch in range(self.num_epochs):
            # predict
            y_predict = self.predict(self.X_data)
            # compute loss
            loss      = self.compute_loss(y_predict,self.y_target)
            self.losses.append(loss)
            # compute gradient 
            k = 2 * ( y_predict - self.y_target )
            gradient = self.X_data.T.dot(k)/self.num_samples
            # Update thetas 
            self.theta = self.theta - self.learning_rate*gradient
            print(f'Epoch: {epoch} - Loss: {loss}')
        return {
            'loss': sum(self.losses)/len(self.losses),
            'weight': self.theta
        }
            
```

```python
def r2score(y_pred, y):
    rss = np.sum((y_pred - y) ** 2)
    tss = np.sum((y-y.mean()) ** 2)
    r2 = 1 - (rss / tss)
    return r2
```

```python
y_pred = np. array ([1 , 2, 3, 4, 5])
y = np. array ([1 , 2, 3, 4, 5])
r2score (y_pred , y)
```

## 3.1.Polymial Regression with degree = 2

```python
def create_polynomial_features (X, degree =2):
    """ Creates the polynomial features
    Args :
    X: A array tensor for the data .
    degree : A intege for the degree of
    the generated polynomial function .
    """
    X_new = X
    for i in range(2,degree+1):
        power_data = np.power(X,i)
        X_new = np.hstack((X_new,power_data))
    return X_new
```

```python
print(X)
# X.re
# Y = np.reshape(X,())
X = X.values.reshape(-1,1)

X_poly = create_polynomial_features(X, degree=2)
X_poly
```

```python
X_poly = np.hstack((np.ones((X_poly.shape[0], 1)), X_poly))
X_poly
```

```python
poly_model = LinearRegression(X_poly, y, learning_rate=0.0001, num_epochs=10000)
poly_model.fit()
```

```python
y_pred = poly_model.predict(X_poly)
r2score(y_pred, y)
```
