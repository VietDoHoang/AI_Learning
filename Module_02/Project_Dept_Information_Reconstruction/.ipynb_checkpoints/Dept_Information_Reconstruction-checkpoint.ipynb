```python
def l1_distance(x,y):
  return abs(x-y)
def l2_distance(x,y):
  return (x-y)**2
def cosine_similarity(x,y):
  numerator = np.dot(x,y)
  denomirator = np.linalg.norm(x)*np.linalg.norm(y)
  return numerator/denomirator
```

```python
def window_based_matching(left_img,right_img,disparity_range=16, compute_type='l1', kernel_size = 3):
  left  = cv2.cvtColor(left_img,cv2.COLOR_BGR2GRAY)
  left  = left.astype(np.float32)

  right = cv2.cvtColor(right_img,cv2.COLOR_BGR2GRAY)
  right  = right.astype(np.float32)

  height,width = left.shape
  disparity = np.zeros((height,width),dtype=np.uint8)
  print(height,width)
  kernel_half = int((kernel_size/2))
  k=0

  if compute_type=='l1':
    max_value = 255*kernel_size*kernel_size
  elif compute_type=='l2':
    max_value = (255**2)*kernel_size*kernel_size
  elif compute_type=='cosine_similarity':
    max_value = -1

  print(kernel_half,height-kernel_half)
  print(kernel_half,width-kernel_half)

  for y in range(kernel_half,height-kernel_half):
    for x in range(kernel_half,width-kernel_half):

      min_cost =max_value
      argmin = 0

      for i in range(disparity_range):

        if x-i-kernel_half<0:
          total_cost = max_value
        else :
          if compute_type =='l1':

            cost = l1_distance(left[y-kernel_half:y+kernel_half+1,x-kernel_half:x+kernel_half+1]
                               , right[y-kernel_half:y+kernel_half+1,x-i-kernel_half:x-i+kernel_half+1])
            total_cost = np.sum(cost)

          elif compute_type=='l2':
            cost = l2_distance(left[y-kernel_half:y+kernel_half+1,x-kernel_half:x+kernel_half+1]
                               , right[y-kernel_half:y+kernel_half+1,x-i-kernel_half:x-i+kernel_half+1])
            total_cost = np.sum(cost)
          elif compute_type=='cosine_similarity':
            wp  = left[y-kernel_half:y+kernel_half+1,x-kernel_half:x+kernel_half+1]
            wqd = right[y-kernel_half:y+kernel_half+1,x-i-kernel_half:x-i+kernel_half+1]
            wp_flattened = wp.flatten()
            wqd_flattened = wqd.flatten()
            cost = cosine_similarity(wp_flattened,wqd_flattened)

        if compute_type=='cosine_similarity':
          if cost >min_cost:
            min_cost = cost
            argmin = i
        else :
          if total_cost < min_cost :
            min_cost = total_cost
            argmin = i
      disparity[y][x] = argmin
  scale = int(255/disparity_range)
  dept_image = disparity*scale
  return dept_image

# def window_based_matching_2(left_img,right_img,disparity_range=16, compute_type='l1', kernel_size = 3):
#   left  = cv2.cvtColor(left_img,cv2.COLOR_BGR2GRAY)
#   left  = left.astype(np.float32)

#   right = cv2.cvtColor(right_img,cv2.COLOR_BGR2GRAY)
#   right  = right.astype(np.float32)

#   height,width = left.shape
#   disparity = np.zeros((height,width),dtype=np.uint8)
#   print(height,width)
#   kernel_half = int((kernel_size/2))
#   k=0

#   if compute_type=='l1':
#     max_value = 255*kernel_size*kernel_size
#   elif compute_type=='l2':
#     max_value = (255**2)*kernel_size*kernel_size

#   print(kernel_half,height-kernel_half)
#   print(kernel_half,width-kernel_half)

#   for y in range(kernel_half,height-kernel_half):
#     for x in range(kernel_half,width-kernel_half):

#       min_cost =max_value
#       argmin = 0
#       # print(min_cost)

#       for j in range(disparity_range):
#         total = 0
#         value = 0
#         for v in range(-kernel_half, kernel_half+1):
#           for u in range(-kernel_half, kernel_half+1):
#             if x+u-j <0:
#               value = max_value
#             else:
#               if compute_type =='l1':
#                 value = l1_distance(int(left[y+v,x+u]),int(right[y+v,x+u-j]))
#               elif compute_type=='l2':
#                 value = l1_distance(int(left[y+v,x+u]),int(right[y+v,x+u-j]))
#               total+=value
#         if total < min_cost:
#           argmin=j
#           min_cost = total

#         # print(argmin)
#       disparity[y][x] = argmin
#       # print(abc)
#   scale = int(255/disparity_range)
#   dept_image = disparity*scale
#   return dept_image
```
